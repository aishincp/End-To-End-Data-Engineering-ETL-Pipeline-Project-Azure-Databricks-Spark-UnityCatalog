### Reading Delta Tables

```python
df = spark.read.table("databricks_etl_catalog.bronze.regions")
```

```python
df.display()
```

```python
df = df.drop("_rescued_data")
```

```python
df.write.format("delta")\
    .mode("overwrite")\
    .save("abfss://silver@databricksetldatastorage.dfs.core.windows.net/regions")
```
